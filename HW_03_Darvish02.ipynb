```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
file_name = "diabetes.csv"
df = pd.read_csv(file_name)
df.head()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(5)
```

```python
column = ['Glucose', 'BloodPressure', 'BMI', 'Age', 'Insulin', 'Outcome']
df_selected = df[column]
correlation_matrix = df_selected.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
outcome_correlation = correlation_matrix['Outcome'].drop('Outcome')
max_correlation_outcome = outcome_correlation.idxmax()
max_correlation_value = outcome_correlation.max()
print(f"The outcome with the highest correlation is: {max_correlation_outcome}")
print(f"The correlation value is: {max_correlation_value}")
```
