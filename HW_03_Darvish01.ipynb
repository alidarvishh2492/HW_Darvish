```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
cities = ('Tehran', 'Mashhad', 'Isfahan', 'Tabriz', 'Shiraz', 'Ahvaz')
days = np.arange(1, 31)
temperature = np.random.randint(15, 40, size=(30, 6))
humidity = np.random.randint(20, 80, size=(30, 6))
rainfall = np.random.randint(0, 50, size=(30, 6))
```

```
data = []
for i in range(len(cities)):
    for day in days:
        data.append({
            'City': cities[i],
            'Day': day,
            'Temperature': temperature[day-1, i],
            'Humidity': humidity[day-1, i],
            'Rainfall': rainfall[day-1, i]
        })
```

```
df = pd.DataFrame(data)
print(df.head())
```

```
mean_city = df.groupby('City')[['Temperature', 'Humidity', 'Rainfall']].mean().round(2)
print(mean_city)
```

```
warmest_city = mean_city['Temperature'].idxmax()
coldest_city = mean_city['Temperature'].idxmin()
print(f"The warmest city is {warmest_city}")
print(f"The coldest city is {coldest_city}")
```

```
rainy_days = df[df['Rainfall'] > 10].groupby('City')['Day'].count()
print(rainy_days)
```

```
isfahan_data = df[df['City'] == 'Isfahan']
plt.figure(figsize=(10, 8))
plt.scatter(isfahan_data['Temperature'], isfahan_data['Humidity'])
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Humidity (%)')
plt.title('Isfahan Temperature vs Humidity in 30 days')
plt.show()
print("\n")
print("As the temperature increases, the humidity also increases.")
```
